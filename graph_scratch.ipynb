```python
G = nx.Graph()
```

```python
G.add_nodes_from([1,2,3,4,5,6,7,8,9])
```

```python
G.add_edges_from([(1,2), (2,3),(3,4),(4,5),(5,8),(8,1),(8,3),(8,7),
                  (9,6),(8,2),(9,7),(9,8),(6,4)])
```

```python
nx.draw(G, with_labels=True, font_weight='bold');
```
